```python
from langchain.tools import tool

@tool
def add_numbers (a: float, b: float) -> float:
    """Perform addition on two real numbers."""
    print("Invoking addition tool...")
    return a + b
```

```python
from langchain.agents import create_agent

agent = create_agent(
    model="ollama:qwen3:4b",
    tools=[add_numbers],
    system_prompt="You are an expert mathematician"
)
```

```python
result = agent.invoke(
    {"messages": [{"role": "user", "content": "What is 3 + 9"}]}
)
for msg in result["messages"]:
    msg.pretty_print()
```
