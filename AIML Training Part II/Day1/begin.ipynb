```python
import pandas as pd
import numpy as np

titanic_df = pd.read_csv("./titanic.csv")
survived_sex_df = titanic_df[['Survived','Sex']]

unsurvived = survived_sex_df[survived_sex_df['Survived'] == 0]
unsurvived_male = unsurvived[unsurvived['Sex'] == 'male']
unsurvived_female = unsurvived[unsurvived['Sex'] == 'female']

survived = survived_sex_df[survived_sex_df['Survived'] == 1]
survived_male = survived[survived['Sex'] == 'male']
survived_female = survived[survived['Sex'] == 'female']


```

```python
import matplotlib.pyplot as plt

x = ['survived_male', 'survived_female', 'unsurvived_male', 'unsurvived_female']
y = [len(survived_male), len(survived_female), len(unsurvived_male), len(unsurvived_female)]
plt.bar(x,y)
plt.show()
```
